```python
import numpy as np

x = np.linspace(0, 10, 30).reshape(-1, 1)
noise =  np.random.normal(0, 1, 30).reshape(-1, 1)
y = x ** 2 + noise
```

```python
from matplotlib import pyplot as plt
plt.scatter(x, y)
plt.xlabel('$X$', fontsize=20)
plt.ylabel('$Y$', fontsize=20)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor

linear_regression = LinearRegression()
decision_tree_regressor = DecisionTreeRegressor(max_depth=3)
k_neighbors_regressor = KNeighborsRegressor(n_neighbors=3)

linear_regression.fit(x, y)
decision_tree_regressor.fit(x, y)
k_neighbors_regressor.fit(x, y)
```

```python
plt.scatter(x, y, s=100)
x_plot = np.linspace(0, 10, 1000).reshape(-1, 1)
plt.plot(x_plot, linear_regression.predict(x_plot), color='red', linewidth=5, alpha=0.8, label='Linear Regression')
plt.plot(x_plot, decision_tree_regressor.predict(x_plot), color='green', linewidth=5, alpha=0.8, label='Decision Tree Regressor')
plt.plot(x_plot, k_neighbors_regressor.predict(x_plot), color='orange', linewidth=5, alpha=0.8, label='K Neighbors Regressor')
plt.xlabel('$X$', fontsize=20)
plt.ylabel('$Y$', fontsize=20)
plt.show()
```

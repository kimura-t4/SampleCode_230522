```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
rng = np.random.default_rng()
x = np.linspace(0, 10, 30).reshape(-1, 1)
y = x ** 2 + rng.normal(0, 1, 30).reshape(-1, 1)

plt.scatter(x, y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
linear_regression = LinearRegression()
decision_tree = DecisionTreeRegressor(max_depth=3)
k_neighbors_regressor = KNeighborsRegressor(n_neighbors=3)
```

```python
linear_regression.fit(x, y)
decision_tree.fit(x, y)
k_neighbors_regressor.fit(x, y)

plt.scatter(x, y, s=100)
x_plot = np.linspace(0, 10, 1000).reshape(-1, 1)
plt.plot(x_plot, linear_regression.predict(x_plot), color='red', linewidth=5, alpha=0.8)
plt.plot(x_plot, decision_tree.predict(x_plot), color='green', linewidth=5, alpha=0.8)
plt.plot(x_plot, k_neighbors_regressor.predict(x_plot), color='orange', linewidth=5, alpha=0.8)
plt.show()
```

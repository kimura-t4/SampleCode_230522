```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
np.random.seed(39)
x = np.linspace(0, 1, 10)
y = 0.5 * x + 0.2 + np.random.randn(len(x)) * 0.03
```

```python
x_plot = np.linspace(0, 1, 100)
plt.scatter(x, y)
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.xlabel(r"$x$", fontsize=20)
plt.ylabel(r"$y$", fontsize=20)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
linear_regression.fit(x.reshape(-1, 1), y.reshape(-1, 1))
```

```python
x_plot = np.linspace(0, 1, 100)
plt.plot(x_plot, linear_regression.predict(x_plot.reshape(-1, 1)), color="orange", label=r"y=0.5x+0.2")
plt.scatter(x, y)
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.xlabel(r"$x$", fontsize=20)
plt.ylabel(r"$y$", fontsize=20)
plt.legend()
plt.show()
```
